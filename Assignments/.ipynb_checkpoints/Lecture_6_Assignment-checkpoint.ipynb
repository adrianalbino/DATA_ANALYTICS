```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('dataset.csv')
data
```

```python
correlation_matrix = data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
df = data

median_price = df['Y house price of unit area'].median()
df['target_label'] = (df['Y house price of unit area'] > median_price).astype(int)

correlation_with_label = df.corr()['target_label'].abs().sort_values(ascending=False)

correlation_with_label = correlation_with_label.drop('target_label')

top_correlated_features = correlation_with_label.head(5)

print(top_correlated_features)
```

Our approach here will be one of using Cross Validation on 90% of the dataset, and then judging our results on a final test set of 10% to evaluate our model.

**TASK: Split the data into features and labels, and then split into a training set and test set, with 90% for Cross-Validation training, and 10% for a final test set.**

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt

# Splitting
X = df.drop(['Y house price of unit area', 'target_label'], axis=1)
y = df['target_label']


X_train_cv, X_test, y_train_cv, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

**TASK: Create a PipeLine that contains both a StandardScaler and a KNN model**

```python
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('knn', KNeighborsClassifier())
])
```

**TASK: Perform a grid-search with the pipeline to test various values of k and report back the best performing parameters.**

```python
param_grid = {
    'knn__n_neighbors': [3, 5, 7, 9, 11]
}


grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train_cv, y_train_cv)


print("Best performing parameters:")
print(grid_search.best_params_)
```

```python
**(HARD) TASK: Using the .cv_results_ dictionary, see if you can create a plot of the mean test scores per K value.**
```

```python
mean_test_scores = grid_search.cv_results_['mean_test_score']
k_values = param_grid['knn__n_neighbors']

plt.figure(figsize=(10, 6))
plt.plot(k_values, mean_test_scores, marker='o')
plt.title('Mean Test Scores per K value')
plt.xlabel('K value')
plt.ylabel('Mean Test Score')
plt.grid(True)
plt.show()
```

**(HARD) TASK: Using the .cv_results_ dictionary, see if you can create a plot of the mean test scores per K value.**

```python
from sklearn.metrics import classification_report, confusion_matrix

y_pred = grid_search.predict(X_test)

print("Classification Report:")
print(classification_report(y_test, y_pred))

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```
