Pearson Correlation Test:

Null Hypothesis (H0): There is no correlation between the number of convenience stores nearby and house prices.

Alternative Hypothesis (H1): There is a correlation between the distance to the number of convenience stores nearby and house prices.

```python
# Extract data from the DataFrame
distance_to_mrt = df['X4 number of convenience stores'].tolist()
house_prices = df['Y house price of unit area'].tolist()

# Calculate the mean of each list
mean_distance = sum(distance_to_mrt) / len(distance_to_mrt)
mean_price = sum(house_prices) / len(house_prices)


# Calculate deviations and covariance
covariance = sum((mrt - mean_distance) * (price - mean_price) for mrt, price in zip(distance_to_mrt, house_prices))
covariance /= (len(distance_to_mrt) - 1)

# Calculate standard deviations
std_distance = (sum((mrt - mean_distance) ** 2 for mrt in distance_to_mrt) / (len(distance_to_mrt) - 1)) ** 0.5
std_price = (sum((price - mean_price) ** 2 for price in house_prices) / (len(house_prices) - 1)) ** 0.5

# Calculate Pearson correlation coefficient
correlation_coefficient = covariance / (std_distance * std_price)

print("Pearson Correlation Coefficient:", correlation_coefficient)
```

Interpretation:

A Pearson Correlation Coefficient of 0.57 shows that there is a moderate or slight positive correlation between the number of convenience stores nearby and the house price of unit area. This means we can possibly reject the null hypothesis.
